```python
import pandas as pd

from sklearn.tree import DecisionTreeClassifier

dados = {
    "horasestudo": [1,2,3,4,5,6],
    "resultado": [0, 0, 0, 1, 1, 1],
    "aprovado": [0,0,0,1,1,1]
}

# Cria a tabela
df = pd.DataFrame(dados)

# Entradas
X = df[["horasestudo", "resultado"]]

# Saídas
y = df["aprovado"]

# Cria o modelo
modelo = DecisionTreeClassifier()

# Treina o modelo
modelo.fit(X, y)

# Nova pessoa
nova_pessoa = [[2.5,1]]

# Faz previsão
resultado = modelo.predict(nova_pessoa)

if resultado == 0:
  print("reprovado")
else:
  print("aprovado")
```
